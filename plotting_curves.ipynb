```python
import numpy as np 
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
x = np.linspace(np.pi, 8*np.pi,50)
plt.plot(x, np.sin(x))
```

```python
def get_noise(shape, mean = 0.5):
    return np.random.normal(0, mean, shape)
```

```python
noisy_x = np.linspace(np.pi, 8*np.pi, 20)
# noise = get_noise(noisy_x.shape)
noise = get_noise(20)

plt.plot(noisy_x, noise)
```

```python
noisy_xt = np.linspace(4*np.pi, 8*np.pi, 10)
full_x = np.concatenate((noisy_x , x, noisy_xt))
full_y = np.concatenate((get_noise(noisy_x.shape), np.sin(x), get_noise(noisy_xt.shape)))
plt.plot(full_x, full_y)
```

```python
full_x = np.concatenate((noisy_x , x, ))
full_y = np.concatenate((np.tanh(noisy_x)+noise, np.sin(x)))
plt.plot(full_x, full_y)
```
